```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.head(10)
```

```python
df.info()
```

#Data Cleaning

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df[['Age','Orders','Amount']].describe()
```

#Data Viz

```python
ax=sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df,palette='Set1')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen= df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales_gen,palette='rocket')
```

From The above graph we can see that the females tend to spend more than males

```python
sns.set(rc={'figure.figsize':(10,7)})
ax=sns.countplot(x='Age Group',data=df,palette='Set1' ,hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age= df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age,palette='rocket')
```

From the above graph we can see that the max amount spent is by the age group 26-35

```python
sales_st= df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,15)},font_scale=1)

sns.barplot(x='State',y='Orders',data=sales_st,palette='rocket')
```

We can see above that UP has the highest order rates and it descreases as we move from left to right on the scale

```python
sales_st= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,15)},font_scale=1)

sns.barplot(x='State',y='Amount',data=sales_st,palette='rocket')
```

We can see that the amount spend is also high in UP. An interesting obvservation is that even though kerala is ranked at 8th position in orders but the amount spend is less as compared to rest States to the right

```python
sns.set(rc={'figure.figsize':(25,9)},font_scale=1)

ax=sns.countplot(data=df,x='Occupation')
for bars in ax.containers:
  ax.bar_label(bars)
```

There by most orders are from IT then healthcare .

```python
sales_st= df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,15)},font_scale=1)

sns.barplot(x='Product_Category',y='Amount',data=sales_st,palette='rocket')
```

Thereby the sold categories are Food ,Clothing and Gadgets.

#Conclusion

Thereby we can say that women who lie in the age range of 25-35 working mostly in IT,Healthcare and Aviation are more likely to buy products from the categories such as Food , Clothing and Gadgets.
